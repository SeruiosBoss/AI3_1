Напишите скрипт для подсчета количества комплексных элементов
в списке, состоящем из элементов произвольного типа

````python
def count_complex_elements(lst):
    """Подсчитывает количество комплексных элементов в списке. """
    return sum(1 for element in lst if type(element) is complex)
````

# Пример использования

````python
example_list = [1, 2.5, 3 + 4j, "text", 5 - 2j, 7, [1, 2], (3, 4), 9j]
complex_count = count_complex_elements(example_list)
print(f"Количество комплексных чисел в списке: {complex_count}")
````

5. **Объясните механизм обращения к элементам списка.**  
- К элементам списка можно обращаться по индексу:  
  - Индексация начинается с 0 для первого элемента.  
  - Отрицательные индексы позволяют обращаться к элементам с конца списка.  
  ```python
  my_list = [10, 20, 30]
  print(my_list[0])  # 10
  print(my_list[-1]) # 30
  ```
- Списки также поддерживают срезы:  
  ```python
  my_list = [1, 2, 3, 4, 5]
  print(my_list[1:4])  # [2, 3, 4]
  ```

6. **Поясните, чем список отличается от массива (например, в C++).**  
- **Гибкость типов данных:**  
  В Python список может содержать элементы разных типов, а в C++ массивы строго типизированы (например, массив `int` может содержать только целые числа).  
- **Длина:**  
  В Python список динамический — его размер может изменяться во время выполнения программы. В C++ массивы имеют фиксированный размер (если используются стандартные массивы) или требуют явного управления памятью (динамические массивы).  
- **Функциональность:**  
  Python предоставляет встроенные методы для работы со списками (`append`, `remove`, `sort`, и т.д.), в то время как в C++ для массивов подобные операции требуют дополнительных библиотек или ручной реализации.  
- **Производительность:**  
  Списки в Python более универсальны, но менее производительны, чем массивы в C++ из-за их динамической структуры и поддержки произвольных типов данных.
