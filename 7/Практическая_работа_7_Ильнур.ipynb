## Обозначение функций, чтение `массива данных csv` через `numpy array` и построение `графиков matplotlib`

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = pd.read_csv('p4_data_16.csv')

# Преобразование колонки 'Money' в числовой формат (убираем символ '$')
data['Money'] = data['Money'].replace('[\\$,]', '', regex=True).astype(float)

# Подготовка данных
salaries = data['Money'].values  # Заработная плата
employee_counts = data['Department'].value_counts()  # Количество сотрудников по отделам

# Построение графиков
plt.figure(figsize=(14, 7))

# График 1: Гистограмма заработной платы
plt.subplot(1, 2, 1)
plt.hist(salaries, bins=20, color='skyblue', edgecolor='black')
plt.title('Распределение заработной платы')
plt.xlabel('Заработная плата ($)')
plt.ylabel('Частота')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# График 2: Круговая диаграмма численности сотрудников по отделам
plt.subplot(1, 2, 2)
plt.pie(employee_counts, labels=employee_counts.index, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'orange'])
plt.title('Численность сотрудников по отделам')

# Отображение графиков
plt.tight_layout()
plt.show()
````

В данной работе были использованы библиотеки `numpy, pandas, matplotlib` и модули `pyplot` для работы с массивами данных `csv` и построения `графиков` на их основе

1. **Чем список Python отличается от массива NumPy?**  
- **Списки Python:**  
  - Могут содержать элементы разных типов (гибкость).  
  - Не поддерживают векторные операции.  
  - Медленнее в вычислениях, так как хранятся как ссылки на объекты.  

- **Массивы NumPy:**  
  - Однородны: элементы массива должны быть одного типа.  
  - Поддерживают векторные операции (ускоряют выполнение операций над данными).  
  - Более эффективны в плане памяти и производительности, так как хранятся в непрерывных блоках памяти.  

2. **Варианты создания массива NumPy:**  
```python
import numpy as np

arr_from_list = np.array([1, 2, 3, 4, 5])
```

3. **Выполнение векторных операций с массивами NumPy:**  
```python
# Умножение вектора на число
vec = np.array([1, 2, 3])
scaled_vec = vec * 2

# Сложение векторов
vec1 = np.array([1, 2, 3])
vec2 = np.array([4, 5, 6])
sum_vec = vec1 + vec2

print(scaled_vec, sum_vec, sep="\n")
```

5. **Основные типы диаграмм Matplotlib и функции для их построения:**  
- **Линейный график (`plot`)**:  
  Используется для отображения зависимости одной величины от другой.  
  ```python
  import matplotlib.pyplot as plt
  plt.plot([1, 2, 3], [4, 5, 6])
  plt.show()
  ```
  
- **Гистограмма (`hist`)**:  
  Для отображения распределения данных.  
  ```python
  plt.hist([1, 2, 2, 3, 3, 3, 4], bins=4)
  plt.show()
  ```

- **Диаграмма разброса (`scatter`)**:  
  Используется для отображения точек данных.  
  ```python
  plt.scatter([1, 2, 3], [4, 5, 6])
  plt.show()
  ```

- **Круговая диаграмма (`pie`)**:  
  Для отображения пропорций.  
  ```python
  plt.pie([15, 30, 45, 10], labels=['A', 'B', 'C', 'D'])
  plt.show()
  ```

- **Столбчатая диаграмма (`bar`)**:  
  Для сравнения величин.  
  ```python
  plt.bar(['A', 'B', 'C'], [10, 20, 15])
  plt.show()
  ```
